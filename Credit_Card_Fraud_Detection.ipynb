**Importing Necessary Libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, f1_score, recall_score, precision_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

**Loading the dataset**

```python
data=pd.read_csv("/content/drive/MyDrive/DATASETS/creditcard.csv")
```

**Data Preprocessing**

```python
#Visualizing class distribution
sns.countplot(x=data['Class'])
plt.title('Class Distribution Before Balancing ')
plt.show()
```

```python
X=data.drop('Class',axis=1)
Y=data['Class']
#TrainTestSplit
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print("X_train shape:",X_train.shape)
print("X_test shape:",X_test.shape)
print("Y_train shape:",Y_train.shape)
print("Y_test shape:",Y_test.shape)
```

**Standardizing only Amount and Time because remaining features were already done using PCA**

```python
#STANDARDSCALER
scaler=StandardScaler()
X_train[['Amount','Time']]=scaler.fit_transform(X_train[['Amount','Time']])
X_test[['Amount','Time']]=scaler.transform(X_test[['Amount','Time']])
```

Applying OverSampling technique to trained data. ✅

DO NOT apply OverSampling on test data because Test data simulates real-world, unseen data. ❌

```python
#RandomOverSampling
ros=RandomOverSampler(random_state=42)
X_train_ros,Y_train_ros=ros.fit_resample(X_train,Y_train)
```

```python
#Visualizing class distribution(balanced data)
sns.countplot(x=Y_train_ros)
plt.title('Class Distribution After Balancing ')
plt.show()
```

```python
#Before oversampling
print('Original dataset shape: {}'.format(Counter(Y_train)))
#After oversampling
print('Resampled dataset shape: {}'.format(Counter(Y_train_ros)))
```

Training and testing the model

```python
def model_evaluation(model,name):
  from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
  model.fit(X_train_ros,Y_train_ros)
  Y_pred=model.predict(X_test)
  #Checking the accuracies of training and testing to know whether the model is overfitted or underfitted
  training_accuracy=accuracy_score(model.predict(X_train_ros),Y_train_ros)
  testing_accuracy=accuracy_score(Y_pred,Y_test)
  print(f"----------{name}----------")
  print()
  print(f"Training Accuracy:{training_accuracy}")
  print(f"Testing Accuracy:{testing_accuracy}")
  print()
  print(classification_report(Y_test,Y_pred))
  cm=confusion_matrix(Y_test,Y_pred)
  sns.heatmap(cm,annot=True,fmt='d')
```

*   LOGISTIC REGRESSION

```python
model_lr=LogisticRegression()
model_evaluation(model_lr,"Logistic Regression")
```



*   Random Forest



```python
model_rf=RandomForestClassifier()
model_evaluation(model_rf,"Random Forest")
```



*   XGBoost



```python
model_xgb=XGBClassifier()
model_evaluation(model_xgb,"XGBoost")
```
